Dataset creation

```python
import os
import pyarrow as pa
import pyarrow.parquet as pq
```

```python
# dataset creation
current_folder = globals()['_dh'][0]
dirname = os.path.dirname(current_folder)

# NOTE: change here the name of the directory that contains your corpus
name = 'Donna_1906gen,giu-1907-1908-1909gen,giu-1910-1913-1914-1916_annotated'
dataset_to_explore = f'data/{name}'
image_dir = os.path.join(dirname, dataset_to_explore)

rows = []

for filename in os.listdir(image_dir):
    if filename.lower().endswith((".jpg",".jpeg",".png",".PNG")):
        path = os.path.join(image_dir, filename)
        with open(path, "rb") as f:
            image_bytes = f.read()

        rows.append({
            "image": image_bytes,
            "filename": filename,
        })
```

```python
table = pa.Table.from_pylist(rows)

pq.write_table(
    table,
    f"{name}.parquet",
    compression="zstd"
)
```

```python
# test dataset
table = pq.read_table(f"{name}.parquet")
df = table.to_pandas()

df.head()
```

```python
# Get image bytes back
image_bytes = df.loc[0, "image"]

from PIL import Image
import io

img = Image.open(io.BytesIO(image_bytes))
img.show()

width, height = img.size
print(f"Width: {width}, Height: {height}")
```
