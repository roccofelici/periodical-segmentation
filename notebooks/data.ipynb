```python
import json
import os
import pandas
```

```python
current_folder = globals()['_dh'][0]
dirname = os.path.dirname(current_folder)

# NOTE: change here the name of the dataset
dataset_to_explore = 'data/LetturaSportiva_1912_giu-lug.csv'
filename = os.path.join(dirname, dataset_to_explore)

df = pandas.read_csv(filename)
df.head()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plt.grid(True, zorder=0)
plt.scatter(df['n_images'], df['area_percentage'], color='grey', zorder=2)
plt.xlabel('abs # images')
plt.ylabel('area by images')
plt.show()
```
